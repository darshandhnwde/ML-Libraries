** **Filtering Data**

```python
new_df = df.loc[(df['Attack']>130) & (df['HP']>130)] # use this or operator for or ("|")
new_df = new_df.reset_index(drop = True)
new_df
# here you can use conditions and relational operator
```

```python
# for removing the attribute or element in columns
df.loc[~df['Type 2'].astype(str).str.contains('Poison')]
```

**Conditional Changes**

```python
# modifiy columns individually and multiple
df.loc[df['Total']>500,['Generation','Legendary']]=['G','L']
df
```

**Aggregate Statistics (Groupby)**

```python
df = pd.read_csv('/content/modified.csv')
df
```

```python
# avarage of Attack columns
df.groupby(['Attack']).mean()
```

```python
# getting avarage according to ascending order
df.groupby(['Type 1']).mean().sort_values('Defense', ascending=False)
```

```python
# also use sum()
df.groupby(['Type 1']).sum()
```

```python
#count ()
df.groupby(['Type 1']).count()
```

```python
# for getting single column or more cleare
df['count']=1
df.groupby(['Type 1']).count()['count']
```

**Working with large Dataset**

```python
# getting small chunk of large data so we can analysis more efficient
for df in pd.read_csv('/content/modified.csv', chunksize=4):
  print("CHUNK OF DATA")
  print(df)
```
