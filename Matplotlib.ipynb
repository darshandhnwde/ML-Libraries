```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

**Line Graph**

```python
x=[1,2,3]
y=[2,4,6]
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Just For Example', fontdict={ 'fontsize':16}) # We can add fontname as well

plt.yticks([2,3,4,5,6]) # If we want just int num then we add like this

plt.plot(x,y , color = 'blue',linewidth=2,marker='.',markersize=10 , linestyle='--') # we can change the color , width of line ,marker and linestyle

# use shorthand notation
#fmt = '[color][marker][linestyle]'
#plt.plot(x,y , 'r.--')

plt.show()
```

```python
# two line graph
y=[10,20,30,40,50]
x=[20,40,60,80,100]
plt.xlabel('X')
plt.ylabel('Y')
x2=np.arange(25,105,5)
plt.plot(x2,x2**2,label='x2^2')
plt.plot(x,y , 'r',label='x')
plt.legend()
plt.show()
```

```python
# resize your graph
plt.figure(figsize=(3,5),dpi=100)
y=[10,20,30,40,50]
x=[20,40,60,80,100]
plt.xlabel('X')
plt.ylabel('Y')

# select interval according to yourself
x2=np.arange(25,105,5)
plt.plot(x2,x2**2,label='x2^2')
plt.plot(x,y , 'r',label='x')
plt.legend()
plt.show()

# to save the graph
plt.savefig('mlgraph.jpg',dpi=200)#you can choose dpi according to your need at the time of saving
```

**Bar Chart**

```python
# remember labels and values have to same argumets
labels=['china', 'india', 'Germany']
values=[1.3,1.4,0.2]
plt.figure(figsize=(7,4))

# title
plt.title('Population in Billions')

# changing the size of grraph
bars = plt.bar(labels, values , color='y')

# coloring and shading bar
patterns = ['','/','*']
for bar in bars:
  bar.set_hatch(patterns.pop(0))


plt.show()
```
